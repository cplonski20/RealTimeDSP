```python

import cv2
import matplotlib.pyplot as plt

img = cv2.imread('logo.png')

plt.figure()
# OpenCV image channel is BGR so we flip the channels to RGB
plt.imshow(img[:,:,::-1])
plt.show()
```

```python
cv2.rectangle(img,(115,50),(135,70),(255, 0,0), thickness=4)
cv2.putText(img,"Prelab7", (40,40),cv2.FONT_HERSHEY_SIMPLEX,1, color=(0,0,255))
plt.imshow(img[:,:,::-1])
plt.show()
```
