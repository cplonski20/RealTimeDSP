```python
import numpy as np
from numpy.fft import fft
import matplotlib.pyplot as plt
import scipy.io.wavfile as spwav
#from mpldatacursor import datacursor
import sys

plt.style.use('ggplot')

# Note: this epoch list only holds for "test_vector_all_voiced.wav"
epoch_marks_orig = np.load("test_vector_all_voiced_epochs.npy")
F_s, audio_data = spwav.read("test_vector_all_voiced.wav")
N = len(audio_data)

######################## YOUR CODE HERE ##############################
F_new = 420
new_epoch_spacing = F_s//F_new

audio_out = np.zeros(N)

# Suggested loop
for i in range(0, N , new_epoch_spacing): # n**2 due to runtime of np.argmin

    # https://courses.engr.illinois.edu/ece420/lab5/lab/#overlap-add-algorithm
    # Your OLA code here

    origidx = np.argmin(np.abs(epoch_marks_orig - i)) #need to change when doing C++ and make quick
    center = epoch_marks_orig[origidx]
    dist = 0
    if(origidx == 0 or origidx >= epoch_marks_orig.shape[0]-1):
        dist = 216
    else:
        dist = ((epoch_marks_orig[origidx+1] - epoch_marks_orig[origidx]) + (epoch_marks_orig[origidx] - epoch_marks_orig[origidx-1]))/2
    
    if(center-dist <0 or center +dist > audio_data.shape[0] or i-dist < 0 or i+dist > len(audio_data)):
        continue
    
    window = np.hanning(2*dist + 1) * audio_data[int(center - dist): int(center + dist +1)]
    audio_out[int(i - dist):int(i + dist + 1)] += window
```

```python
from IPython.display import Audio
from scipy.io.wavfile import read, write

Audio(audio_out, rate=F_s)
```
