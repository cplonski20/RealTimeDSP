```python
import numpy as np
from numpy.fft import fft
import matplotlib.pyplot as plt
import scipy.io.wavfile as spwav
from IPython.display import Audio
#from mpldatacursor import datacursor
import sys

plt.style.use('ggplot')

# Note: this epoch list only holds for "test_vector_all_voiced.wav"
epoch_marks_orig = np.load("test_vector_all_voiced_epochs.npy")
F_s, audio_data = spwav.read("test_vector_all_voiced.wav")
N = len(audio_data)

######################## YOUR CODE HERE ##############################
F_new = 420
new_epoch_spacing = F_s//F_new

audio_out1 = np.zeros(N)

for i in range(0, N, new_epoch_spacing):
    closestEpoch = np.argmin(abs(epoch_marks_orig-i))
    if closestEpoch+1 < len(epoch_marks_orig) and closestEpoch-1 > -1:
        P = (epoch_marks_orig[closestEpoch+1] - epoch_marks_orig[closestEpoch-1])/2
        windowLength = 2 * P + 1
        window = np.hanning(windowLength)
        impulseResponse = window * audio_data[epoch_marks_orig[closestEpoch-1]:epoch_marks_orig[closestEpoch+1]+1]
        audio_out1[int(i - P):int(i + P) + 1] += impulseResponse
    elif closestEpoch+1 >= len(epoch_marks_orig):
        P = (epoch_marks_orig[closestEpoch] - epoch_marks_orig[closestEpoch - 1])
        windowLength = P
        window = np.hanning(windowLength)
        impulseResponse = window * audio_data[epoch_marks_orig[closestEpoch - 1]:epoch_marks_orig[closestEpoch]]
        audio_out1[int(i - P):int(i)] += impulseResponse
        print("right edge")
    elif closestEpoch-1 <= -1:
        P = (epoch_marks_orig[closestEpoch + 1] - epoch_marks_orig[closestEpoch])
        windowLength = P
        window = np.hanning(windowLength)
        impulseResponse = window * audio_data[epoch_marks_orig[closestEpoch]:epoch_marks_orig[closestEpoch + 1]]
        audio_out1[int(i):int(i + P)] += impulseResponse
        print("left edge")

Audio(audio_out1, rate = F_s)
```
